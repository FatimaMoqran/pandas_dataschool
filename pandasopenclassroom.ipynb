```python
import numpy as np
import pandas as pd
```

# Cardinalité un à un

```python
df1= pd.DataFrame({'employee':['Bob','Jake','Lisa','Sue'],
                  'department':['Accounting','Engineering','Engineering','HR']})
df2 = pd.DataFrame({'employee':['Lisa', 'Bob', 'Jake', 'Sue'],
                  'date':[2004,2008,2012,2014]})
df3 = pd.merge(df1,df2)

df3
```

# Cardinalité plusieurs à plusieurs

```python
df5 = pd.DataFrame({'department':['Accounting','Accounting','Engineering','Engineering','HR','HR'],
                    'competence':['math','spreadsheets','coding','linux','spreadsheets','organisation']})
```

```python
pd.merge(df1,df5)
```

# Le produit cartesien

```python
#on ajoute une nouvelle colonne à df1 et df2 contient toujours la même valeur
df1['key'] = 0
df2['key'] = 0

#la jointure plusieurs à plusieurs
produit_cartesien = pd.merge(df1,df2, on='key')

```

```python
produit_cartesien
```

```python
produit_cartesien.drop('key',1,inplace=True)
```

```python
pd.merge(df1.assign(key=0),df2.assign(key=0),on='key').drop('key',axis=1)
```

## L'agrégation

```python
rng = np.random.RandomState(42)
ser = pd.Series(rng.rand(5))
print(ser.sum())
print(ser.mean())
```

```python
df =pd.DataFrame({'A':rng.rand(5),
                 'B':rng.rand(5)})
#Par colonne
print(df.mean())

#Par ligne
print(df.mean(axis='columns'))
```

```python
df = pd.DataFrame({'key': ['A', 'B', 'C','A','B','C'],
                    'data1':range(6),
                   'data2': [10,11,10,9,10,10]})
print(df)
```

```python
#on crée un objet de type DataFrameGroupBy que l'on va appeler gb
gb =df.groupby('key')
print(gb.sum())
print(gb.mean())
```

```python
s = gb['data1','data2'].sum()
m = gb['data2',].mean()

groupped = pd.concat([s,m], axis=1)
groupped.columns = ["data1_somme","data2_somme","data2_moyenne"]

groupped
```
